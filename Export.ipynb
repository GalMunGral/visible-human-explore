```python
import os
import numpy as np
from pydicom import dcmread
import matplotlib.pyplot as plt
from mpl_interactions import ipyplot as iplt
from PIL import Image
```

```python
data_dir = './data'

for i in range(len(os.listdir(data_dir))):
    filepath = os.path.join(data_dir, f'VHFCT1mm-Head ({i + 1}).dcm')
    ds = dcmread(filepath)
    im = Image.fromarray(ds.pixel_array * 255)
    im.save(f"texture-{i}.png")
```
