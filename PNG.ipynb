```python
import os
import numpy as np
from pydicom import dcmread
import matplotlib.pyplot as plt
from mpl_interactions import ipyplot as iplt
from PIL import Image
```

```python
data_dir = './data'
os.makedirs('./out', exist_ok=True)

n = len(os.listdir(data_dir))

slices = []
max_value = 0

for i in range(n):
    filepath = os.path.join(data_dir, f'VHFCT1mm-Head ({i + 1}).dcm')
    ds = dcmread(filepath)
    dtype = ds.pixel_array.dtype
    max_value = max(max_value, ds.pixel_array.max())
    slices.append(ds.pixel_array)
    
for i in range(n):
    im = Image.fromarray((slices[i] / max_value * 255).astype(np.uint8))
    im.save(f"./out/texture-{i}.png")
```
