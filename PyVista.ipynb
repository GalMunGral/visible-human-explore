```python
import os
import pyvista as pv
import numpy as np
```

```python
data_dir = './data'

slices = []
for i in range(len(os.listdir(data_dir))):
    reader = pv.DICOMReader(f'./data/VHFCT1mm-Head ({i + 1}).dcm')
    dataset = reader.read()
    slices.append(dataset.point_data['DICOMImage'])
    
vol = pv.ImageData(dimensions=(512, 512, 234))
vol.point_data['value'] = np.concat(slices)
```

```python
pv.set_jupyter_backend('client')
pl = pv.Plotter()
pl.add_volume(vol, cmap='viridis', opacity=[.0, .0, .0, .01, .1, .1, .0, .0, .0, .0, .0], shade=True)
pl.show()
```
